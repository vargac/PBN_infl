```python
#attractors should either be {str:set} (fixed point) or {tuple of str:list of sets} (limit cycle)
dattractors={}
names=data['names']
for attractor in data['attractors']:
    #print(attractor)
    if len(attractor)>1: #limit cycle
        key=tuple(''.join([str(int(s)) for s in config]) for config in attractor)
        value=[{names[i]+'-'+str(int(s)) for i,s in enumerate(config)} for config in attractor]
    else:
        key=''.join([str(int(s)) for s in attractor[0]])
        value={names[i]+'-'+str(int(s)) for i,s in enumerate(attractor[0])}
    #print(key,value)
    dattractors[key]=value
len(dattractors),dattractors
```

```python
data['name']
with open('attractors/'+data['name']+'.json','w') as fp:
    fp.write(str(dattractors))
```
